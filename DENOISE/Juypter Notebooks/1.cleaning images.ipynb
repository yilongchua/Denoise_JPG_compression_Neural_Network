```python
import glob
import pandas
from skimage import io
import pandas as pd
import numpy as np
import skimage
import matplotlib.pyplot as plt
import cv2
import math
from PIL import Image
```

```python
path = 'D:/YL/Images/'
images = list(glob.glob(path + '/*JPG'))
df = pd.DataFrame(images,  columns=['image_path'])
df['num'] = df.image_path.apply(lambda x : int(x.split('.')[0].split('_')[1]))
df.sort_values('num', inplace = True)
df.reset_index(drop=True, inplace = True)
df
```

# Resize to (1024,1024,3) Image shape

```python
save_path = 'D:/YL/clean_images/'
for i in range(df.shape[0]):
# for i in range(2):

    image_path = df.image_path[i]
    name = image_path.split('Images')[1][1:]
    image = cv2.imread(df.image_path[i])
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB )
    image_resize = cv2.resize(image, (1024, 1024), interpolation=cv2.INTER_AREA)
    
    cv2.imwrite( save_path + name,cv2.cvtColor(image_resize,cv2.COLOR_RGB2BGR ))
```

# Create Noise in image using Gaussian filter under skimage

```python
save_path = 'D:/YL/noise_images/'
for i in range(df.shape[0]):
    image_path = df.image_path[i]
    name = 'noisy_'+image_path.split('clean_images')[1][1:]
#     print(name)
    origin = skimage.io.imread(image_path)
    noisy = skimage.util.random_noise(origin, mode='gaussian', var=0.9, seed=27, clip=True)
    skimage.io.imsave(save_path+name,noisy, quality=100)
```

# Image split into tiles

```python
def image_split(img_path, kernal_size=(256,256)):
    size = kernal_size[0]
    img = Image.open(img_path)
    imgwidth, imgheight = img.size
    plt.imshow(img)
    plt.show()
    image_ls =[]
    for row in range(math.ceil(imgwidth/ size)):
        img_row = []
        for col in range(math.ceil(imgheight/size)):
            resize = False
            bott_x = (row+1) * size
            if bott_x > imgwidth:
                bott_x = imgwidth
                resize = True
            bott_y = (col + 1) * size
            if  bott_y > imgheight: 
                bott_y = imgheight
                resize= True
            box = (row * size, col * size, bott_x, bott_y)

            img_crop = img.crop(box)
            if resize:
                img_crop_new = img_crop.resize(kernal_size)
            img_row.append(img_crop)
        image_ls.append(img_row)

    img_col = []
    for i in image_ls:
        img_row = [ np.array(pic) for pic in i ]

        img = cv2.vconcat(img_row)
        img_col.append(img)
    img_complete = cv2.hconcat(img_col)
#     plt.imshow(img_complete)
#     plt.show()
    
    rows = len(image_ls[0]) 
    columns = len(image_ls)
    figsize = [20, 20]    


    fig, ax = plt.subplots(nrows=rows, ncols=columns, figsize=figsize)

    for y in range(columns):
    #     print(y)
        for x  in range(rows):
            image_item = image_ls[y][x]
            ax[x,y].set_title(f'{str(x)}, {str(y)}')  # set title
            ax[x,y].imshow(image_item)

    plt.tight_layout(True)

    plt.show() 
    
    return image_ls
```
